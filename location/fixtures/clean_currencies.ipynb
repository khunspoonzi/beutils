```
import json
from unidecode import unidecode
```

```
def slugify(string, decode=True):
    """ Returns a slugified version of a string input """

    # Return empty string if no text value passed
    if not string:
        return ""

    # Remove special characters
    if decode:
        string = unidecode(string)

    # Lowercase and strip slug
    string = string.lower().strip()

    # Return slugified string
    return string
```

```
with open("countries1.json") as f:
    c1 = json.load(f)
    
with open("countries2.json") as f:
    c2 = json.load(f)
```

```
countries = []

regions = []

subregions = []
_subregions = []
```

```
for cc1 in c1:
    
    iso2 = cc1["iso2"]
    iso3 = cc1["iso3"]

    if iso3 == "XKX":
        cc2 = [c for c in c2 if c["iso3"] == "UNK"][0]
    
    else:
        cc2 = [c for c in c2 if c["iso3"] == iso3][0]
        
    name = cc1["name"]
    name_official = cc2["name_official"]
    
    phone_code = cc1["phone_code"]
    if phone_code.startswith("+"):
        phone_code = phone_code.replace("+", "", 1)
    
    phone_codes = [i.strip() for i in phone_code.split(" and ") if i.strip()]
    
    print(phone_codes)

    region = cc1["region"]
    subregion = cc1["subregion"]
    
    if region not in regions:
        regions.append(region)
    if subregion not in subregions:
        subregions.append(subregion)
    
    _subregion = f"{subregion} || {region}"
    if _subregion not in _subregions:
        _subregions.append(_subregion)
        
    flag_path = cc2["flag_path"]
    flag_path = f"beutils_location/country/flag/{flag_path.split('hjnilsson/')[-1]}"
        
    countries.append(
        {
            "region": region,
            "subregion": subregion,
            "currency_code": cc1["currency"],
            "name": name,
            "name_official": name_official,
            "name_native": cc1["native"] or "",
            "iso2": iso2,
            "iso3": iso3,
            "phone_codes": phone_codes,
            "demonym": cc2["demonym"],
            "is_nationality": cc2["is_nationality"],
            "emoji": cc1["emoji"],
            "emoji_u": cc1["emojiU"],
            "alternate_names": cc2["alternate_spellings"],
            "flag_path": flag_path,
            "capital_name": cc1["capital"],
            "latitude": cc1["latitude"],
            "longitude": cc1["longitude"],
            "timezones": cc1["timezones"],
            "translations": cc1["translations"],
        }
    )
    
    
```

```
countries = sorted(countries, key=lambda x: slugify(x["name"]))
```
